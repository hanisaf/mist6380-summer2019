```python
s = pd.DataFrame( {'cust': ['J', 'J', 'K', 'K', 'J'], 'ontime': [True, True, False, True, False]})
```

```python
s
```

```python
def formula(payments_data):
    basic_score = 600
    for d in payments_data:
        if d:
            basic_score += 10
        else:
            basic_score -= 20
    return basic_score
```

```python
formula([True, True, False])
```

```python
s
```

```python
formula(s[s.cust == 'J'].ontime)
```

```python
formula(s[s.cust == 'K'].ontime)
```

```python
s.groupby('cust').ontime.apply(formula)
```
