```python
import networkx as nx
%matplotlib inline
import pandas as pd
from networkx.algorithms import community
```

Load the Zachary's karate club dataset

W. W. Zachary, An information flow model for conflict and fission in small groups, Journal of Anthropological Research 33, 452-473 (1977).

```python
net = nx.read_gml('../data/karate.gml', label='id')
```

# Network structure

```python
nx.degree_centrality(net)
```

```python
nx.betweenness_centrality(net)
```

```python
nx.clustering(net)
```

```python
degree = pd.Series(nx.degree_centrality(net)).sort_values(ascending=False)
between= pd.Series(nx.betweenness_centrality(net)).sort_values(ascending=False)
close =  pd.Series(nx.closeness_centrality(net)).sort_values(ascending=False)
page = pd.Series(nx.pagerank(net)).sort_values(ascending=False)
clust= pd.Series(nx.clustering(net)).sort_values(ascending=False)
```

# community detection

```python
communities = community.greedy_modularity_communities(net)
```

```python
communities
```

```python
colors = ['r', 'b', 'g', 'y']
next(iter(colors))
```

```python
colors = iter(['r', 'b', 'g', 'y'])
pos = nx.kamada_kawai_layout(net)
nx.draw_networkx_labels(net, pos=pos)
nx.draw_networkx_edges(net, pos=pos)
for community in communities:
    nx.draw_networkx_nodes(net, nodelist=community, pos=pos, node_color=next(colors))
```
