```python
import seaborn as sns
import pandas as pd
```

```python
data= sns.load_dataset("iris")
```

```python
data.head()
```

```python
sns.distplot(data.sepal_length)
```

```python
sns.distplot(data.sepal_length)
sns.distplot(data.sepal_width)
sns.distplot(data.petal_length)
sns.distplot(data.petal_width)
```

```python
sns.kdeplot(data.sepal_length)
sns.kdeplot(data.sepal_width)
sns.kdeplot(data.petal_length)
sns.kdeplot(data.petal_width)
```

```python
sns.relplot(x='sepal_length', y='petal_length', data=data)
```

```python
sns.jointplot(x='sepal_length', y='petal_length', data=data)
```

```python
sns.jointplot(x='sepal_length', y='petal_length', kind='hex', data=data)
```

```python
sns.jointplot(x='sepal_length', y='petal_length', kind='kde', data=data)
```

```python
data.corr()
```

```python
sns.pairplot(data)
```

```python
sns.pairplot(data, hue='species')
```
