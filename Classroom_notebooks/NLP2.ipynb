```python
import pandas as pd
```

```python
#!pip install wordcloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline
```

# Visualize word frequencies using wordcloud

```python
word_frequencies = {'contract':100, 'kitchen': 50, 'Excel': 25, 'SQL' : 25}
```

```python
wc=WordCloud(max_font_size=50, max_words=100, background_color="white")
```

```python
wc.generate_from_frequencies(word_frequencies)
plt.figure(figsize=(20,10))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
```

# We would like to use the same technique to visualize the first five earning calls

```python
data = pd.read_csv('../data/EC2000.csv.gz')
```

```python
data2 = data.drop('index', axis='columns')
```

```python
data3 = data2.rename({'fName':'company'}, axis='columns')
```

```python
last_three_cols = data3.columns[-3:]
```

```python
data4 = data3.drop(last_three_cols, axis='columns')
```

```python
data4 = data4.head(5) #first 5
```

```python
text1 = data4.iloc[0].text
```

```python
words = text1.split()
```

```python
sw = pd.Series(words)
```

```python
sw = sw.str.capitalize()
```

```python
freq_dict = sw.value_counts()
```

```python
wc.generate_from_frequencies(freq_dict)
plt.figure(figsize=(20,10))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
```
