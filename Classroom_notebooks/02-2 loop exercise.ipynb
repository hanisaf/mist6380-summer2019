1- Write a loop that prints numbers from 1 to 31, append the prefix "1-" to the numbers, these are day of January, see sample output below

```python
for day in range(1, 32):
    print( f"1-{day}")
```

2- Vary the prefix from 1- to 12- to print all days of the year, do not worry about incorrect dates for now

```python
month = 1
while month <= 12:
    day = 1
    while day <= 31:
        print( f"{month}-{day}")
        day += 1
    month += 1
```

```python
for month in range(1, 13):
    for day in range(1, 32):
        print( f"{month}-{day}")
```

3- now, address the issue of some months being only 30 days and february if 28 days (this year)

```python
for month in range(1, 13):
    if month == 2:
        for day in range(1, 29):
            print( f"{month}-{day}")
    elif month == 4 or month == 6 or month == 9 or month == 11:
        for day in range(1, 31):
            print( f"{month}-{day}")        
    else:
        for day in range(1, 32):
            print( f"{month}-{day}")
```

```python
for month in range(1, 13):
    endrange = 32
    if month == 2:
        endrange = 29
    elif month == 4 or month == 6 or month == 9 or month == 11:
        endrange = 31

    for day in range(1, endrange):
        print( f"{month}-{day}")
```
