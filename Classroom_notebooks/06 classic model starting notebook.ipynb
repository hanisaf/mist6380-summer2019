# One to many relationship

1. Report the account representative for each customer.


```python
res=(pd.merge(Customers, Employees, left_on='salesRepEmployeeNumber', right_on='employeeNumber', how='inner'))
```

```python
res.to_html('res.html') #exporting
```

```python
res[['customerName', 'firstName', 'lastName']]
```

2. Report total payments for Atelier graphique.

```python
pd.merge(Customers, Payments).groupby('customerName')['amount'].sum()['Atelier graphique']
```

```python
step1 = pd.merge(Customers, Payments)
step2 = step1.groupby('customerName')['amount']
step3 = step2.sum()
step3['Atelier graphique']
```
